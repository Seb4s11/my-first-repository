```python
def continue_game(user_choice):
    if user_choice.upper() in ('Y', 'N'):
        if user_choice.upper() == 'Y':
            return True
    return False

def start_game():
    print("Welcome to Rock, Paper, Scissors")
    user_choice = input("Would you like to play? (Y/N) ")
    return continue_game(user_choice)

def terminate_game():
    user_choice = input("Would you like to play again? (Y/N) ")
    return continue_game(user_choice)

def show_game_status(current_round, user_points, computer_points, current_winner):
    print(f"\nCurrent round: {current_round}")
    print(f"User Points: {user_points}")
    print(f"Computer Points: {computer_points}")
    print(f"Current Winner: {current_winner}\n")

def advance_round(current_round):
    return current_round + 1

def play_game():
    game_status = start_game()
    current_round = 0
    user_points = 0
    computer_points = 0
    current_winner = "---"
    while game_status:
        current_round = advance_round(current_round)
        show_game_status(current_round, user_points, computer_points, current_winner)
        game_status = terminate_game()



if _name_ == "_main_":
    play_game()
```
